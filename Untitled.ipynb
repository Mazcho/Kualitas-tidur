# Regresi

```python
X_duration = df_regresi[["Quality of Sleep","Stress Level","Physical Activity Level","BMI Category","Occupation"]] 
y_duration = df_regresi["Sleep Duration"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_normalisasi = scaler.fit_transform(X_duration)
```

```python
X_train2,X_test2,y_train2,y_test2 = train_test_split(X_normalisasi,y_duration,test_size=0.2,random_state=42)
```

```python
model2.fit(X_train2,y_train2)
y_pred_duration = model2.predict(X_test2)

r2_score_duration_GBR = r2_score(y_test2,y_pred_duration)
rmse_GBR = np.sqrt(mean_squared_error(y_test2,y_pred_duration))

print(r2_score_duration_GBR)
print(rmse_GBR)
```

```python
model4.fit(X_train2,y_train2)
y_pred_duration = model4.predict(X_test2)

r2_score_duration_XGBR = r2_score(y_test2,y_pred_duration)
rmse_XGBR = np.sqrt(mean_squared_error(y_test2,y_pred_duration))

print(r2_score_duration_XGBR)
print(rmse_XGBR)
```

```python
model6.fit(X_train2,y_train2)
y_pred_duration = model6.predict(X_test2)

r2_score_duration_RF = r2_score(y_test2,y_pred_duration)
rmse_RF = np.sqrt(mean_squared_error(y_test2,y_pred_duration))

print(r2_score_duration_RF)
print(rmse_RF)
```

```python
model7.fit(X_train2,y_train2)
y_pred_duration = model7.predict(X_test2)

r2_score_duration_ADA = r2_score(y_test2,y_pred_duration)
rmse_ADA = np.sqrt(mean_squared_error(y_test2,y_pred_duration))

print(r2_score_duration_ADA)
print(rmse_ADA)
```

# Klasifikasi

```python
X_sleep_disorder = df[["Quality of Sleep","Stress Level","Physical Activity Level","TOP Blood Preasure","Bottom Blood Preasure","Occupation"]] 
y_sleep_disorder = df["Sleep Disorder"]
```

```python
from imblearn.over_sampling import SMOTE
balancing = SMOTE(k_neighbors=3)
X_resampled, y_resampled = balancing.fit_resample(X_sleep_disorder,y_sleep_disorder)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_normalisasi_klasifikasi = scaler.fit_transform(X_resampled)
```

```python
X_train1,X_test1,y_train1,y_test1 = train_test_split(X_normalisasi_klasifikasi,y_resampled,test_size=0.2,random_state=42)
```

```python
model1.fit(X_train1,y_train1)
y_pred_disorder = model1.predict(X_test1)

classification_report_gbr = classification_report(y_test1,y_pred_disorder)
f1_scoregbr = f1_score(y_test1,y_pred_disorder,average='macro')

print("GBR Model")
print(classification_report_gbr)
```

```python
model3.fit(X_train1,y_train1)
y_pred_disorder = model3.predict(X_test1)

classification_report_RF = classification_report(y_test1,y_pred_disorder)

print("RF Model")
print(classification_report_RF)
```

```python
model5.fit(X_train1,y_train1)
y_pred_disorder = model5.predict(X_test1)

classification_report_XGBC = classification_report(y_test1,y_pred_disorder)

print("XGBC Model")
print(classification_report_XGBC)
```

```python
model8.fit(X_train1,y_train1)
y_pred_disorder = model8.predict(X_test1)

classification_report_Ada = classification_report(y_test1,y_pred_disorder)

print("AdaBoost Model")
print(classification_report_Ada)
```
